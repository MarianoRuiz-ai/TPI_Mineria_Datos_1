# Notebook 05 — Conclusiones


**Objetivo:** Responder las preguntas definidas al inicio del proyecto con evidencia obtenida durante el análisis, diferenciando evidencia, interpretación y conclusión.

## Pregunta 1: ¿El plan de suscripción influye en el tiempo de visualización mensual?

**Evidencia:** Los usuarios Premium tienen el mayor tiempo de visualización promedio, seguidos por Estándar y Básico. La diferencia es consistente tanto en media como en mediana, descartando que sea resultado de valores extremos.

**Interpretación:** Existe una relación positiva entre el plan de suscripción y el tiempo de visualización: a mayor categoría de plan, mayor consumo promedio. Esta relación podría deberse a que los planes superiores ofrecen más contenido o funcionalidades, incentivando un mayor uso; o bien a que los usuarios con mayor consumo migran hacia planes más completos.

**Conclusión:** Sí. El plan de suscripción se asocia al tiempo de visualización mensual. Los usuarios Premium son los de mayor consumo, y los usuarios Básico los de menor consumo. La relación es gradual y consistente en toda la distribución.

## Pregunta 2: ¿Los usuarios con mayor tiempo de visualización generan menos tickets de soporte?

**Evidencia:** La correlación entre `monthly_watch_time_mins` y `customer_support_tickets` es cercana a cero. El análisis por cuartiles de consumo muestra tickets promedio similares en todos los grupos.

**Interpretación:** No existe una relación lineal entre el tiempo de visualización y la cantidad de tickets de soporte. Esto sugiere que los tickets responden a factores distintos al nivel de consumo, como problemas técnicos, errores de facturación o características del plan, que no se capturan con la variable de visualización.

**Conclusión:** No. Los usuarios con mayor tiempo de visualización no generan menos tickets de soporte. El volumen de tickets es independiente del nivel de consumo dentro de los rangos observados en el dataset.

## Pregunta 3: ¿Existen perfiles diferenciados de usuarios según su comportamiento de consumo y soporte?

**Evidencia:** El análisis multivariado mostró que las variables numéricas tienen correlaciones bajas entre sí, indicando que cada una aporta información independiente al perfil del usuario. La proyección PCA en PC1-PC2 muestra cierta separación entre planes en el eje de consumo (PC1), con superposición parcial entre grupos.

**Interpretación:** Existen perfiles diferenciados en términos de consumo y plan, pero la identidad de cada perfil es multivariada: no basta con una sola variable para describir a un usuario. PC1 captura principalmente la dimensión de consumo y plan, mientras PC2 captura la dimensión demográfica y de soporte.

**Conclusión:** Sí, parcialmente. Existen perfiles diferenciados según el nivel de consumo y el plan de suscripción, pero con superposición entre grupos. El perfil de un usuario está determinado por la combinación de múltiples dimensiones, no por una variable única.

## Limitaciones

- El alcance de las conclusiones se encuentra condicionado por la información disponible y por las decisiones documentadas durante el proceso. El dataset no incluye información sobre el contenido consumido, la duración de las sesiones individuales ni el historial de pagos, lo que limita la profundidad del análisis de perfiles.
- La imputación de `monthly_watch_time_mins` con la mediana y de `favorite_genre` con la moda introduce simplificaciones que pueden suavizar diferencias reales en esas variables.
- La eliminación de registros con fechas futuras (882 registros) reduce la representatividad del dataset, aunque es necesaria para mantener la coherencia temporal.
- El análisis es de carácter exploratorio: las relaciones observadas indican asociación estadística, no causalidad.

## Mejoras futuras

- Una mejora futura podría consistir en aplicar técnicas de clustering (k-means, DBSCAN) sobre el espacio PCA para formalizar la segmentación de usuarios y validar cuantitativamente los perfiles identificados.
- Incorporar variables adicionales como historial de pagos, duración de permanencia en la plataforma o cantidad de sesiones permitiría ampliar el alcance del análisis de perfiles.
- Explorar la relación entre tickets de soporte y tipo de plan o antigüedad del usuario podría explicar los factores que sí determinan el volumen de soporte.
